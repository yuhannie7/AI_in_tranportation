# Assignment task: Find the best neural network model for the bus delay prediction


```python
from json import load
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import random
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from keras.models import load_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

#-------------------------------------------------data preprocessing------------------------------------------------------------------

# Define the URL of a CSV file containing data.
url = 'https://raw.githubusercontent.com/zhenliangma/Applied-AI-in-Transportation/master/Exercise_7_Neural_networks/Exercise7data.csv'

# Read the CSV data from the specified URL into a DataFrame
df = pd.read_csv(url)

# Limit the DataFrame to the first 1000 rows
df = df.iloc[:1000]

# Drop specific columns (Arrival_time, Stop_id, Bus_id, Line_id)
df = df.drop(['Arrival_time', 'Stop_id', 'Bus_id', 'Line_id'], axis=1)

# Extract the features and target variable
x = df.drop(['Arrival_delay'], axis=1)
y = df['Arrival_delay']

# Split the train and test data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#-------------------------------------------------network construction------------------------------------------------------------------

# Define the parameters to vary
nn_params = {
    'n_layers': [2, 4, 6, 8, 12],
    'n_neurons': [16, 32, 64, 128],
    'dropout_rate': [0, 0.2, 0.4, 0.5, 0.6],
}

# Function to construct the neural network model with varying parameters
def construct_network_model(n_layers, n_neurons, dropout_rate):
    model = Sequential()
    model.add(Dense(n_neurons, activation='relu', input_dim=X_train.shape[1]))

    for _ in range(n_layers - 1):
        model.add(Dense(n_neurons, activation='relu'))
        model.add(Dropout(dropout_rate))

    model.add(Dense(1))  # Output layer for regression
    return model

# DataFrame to store results
output_df = pd.DataFrame(columns=['n_layers', 'n_neurons', 'dropout_rate', 'MAE', 'MSE', 'R2'])

# Iterate over 5 random configurations
for i in range(10):
    # Randomly sample parameters
    n_layers = random.choice(nn_params['n_layers'])
    n_neurons = random.choice(nn_params['n_neurons'])
    dropout_rate = random.choice(nn_params['dropout_rate'])

    # Construct and compile the model
    model = construct_network_model(n_layers, n_neurons, dropout_rate)
    model.compile(optimizer='adam', loss='mae', metrics=['mae'])

    # Callbacks for early stopping and learning rate reduction
    early_stop = EarlyStopping(monitor='val_mae', patience=5, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_mae', factor=0.5, patience=3)
    filepath = f"weights.best_{i}.h5.keras"
    checkpoint = ModelCheckpoint(filepath, monitor='val_mae', verbose=1, save_best_only=True, mode='min')

    # Train the model
    hist = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32, callbacks=[early_stop, reduce_lr, checkpoint], verbose=0)

    # Evaluate the model on the test data
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Save results to DataFrame
    output_df.loc[len(output_df)] = [n_layers, n_neurons, dropout_rate, mae, mse, r2]

    # Plot the training and validation loss
    err = hist.history['mae']
    val_err = hist.history['val_mae']
    epochs = range(1, len(err) + 1)

    plt.figure(figsize=(10, 5))
    plt.plot(epochs, err, '-', label='Training MAE')
    plt.plot(epochs, val_err, ':', label='Validation MAE')
    plt.title(f'Training and Validation MAE (Config {i})')
    plt.xlabel('Epoch')
    plt.ylabel('Mean Absolute Error')
    plt.legend()
    plt.show()

# Display the results of all configurations
print(output_df)

# Optionally, you can load the best model saved from one of the configurations and evaluate it again
# For example, loading the model from the first configuration:
model = load_model("weights.best_0.h5.keras")
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('--------------------------------------'+'This is result of the model loaded from the local path'+"---------------------------------------------")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
output_df_sorted = output_df.sort_values(by='MAE', ascending=True)
print(output_df_sorted)
```

```python

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.barplot(x=output_df_sorted.index, y='MAE', data=output_df_sorted, ax=axes[0])
axes[0].set_title('Mean Absolute Error by Model Configuration (Sorted)')
axes[0].set_xlabel('Model Index')
axes[0].set_ylabel('MAE')

sns.barplot(x=output_df_sorted.index, y='MSE', data=output_df_sorted, ax=axes[1])
axes[1].set_title('Mean Squared Error by Model Configuration (Sorted)')
axes[1].set_xlabel('Model Index')
axes[1].set_ylabel('MSE')

sns.barplot(x=output_df_sorted.index, y='R2', data=output_df_sorted, ax=axes[2])
axes[2].set_title('R-squared by Model Configuration (Sorted)')
axes[2].set_xlabel('Model Index')
axes[2].set_ylabel('R-squared')

plt.tight_layout()
plt.show()
```
